```python
#Decision Tree Regressor
# 
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
print(type(diabetes))
for i in diabetes:
  print(type(i),i)
  print(type(diabetes[i]), diabetes[i])
```

```python
print(diabetes['feature_names'])
print(diabetes['data'], diabetes['data'].shape, type(diabetes['data']))
print(diabetes['target'], diabetes['target'].shape, type(diabetes['target'])) 
```

```python
X, y = load_diabetes(return_X_y=True)
print(type(X), X.shape, X)
print(type(y), y.shape, y)
```

```python
regressordt = DecisionTreeRegressor()
output = regressordt.fit(X,y)
output
```

```python
#attributes of DecisionTreeRegressor
print(diabetes['feature_names'])
print(regressordt.feature_importances_) #bmi has the highest value after s5 making it most important contributor to diabetes.
print(regressordt.max_features_)
print(regressordt.n_outputs_)
print(regressordt.tree_)
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(regressordt, X, y)
score #https://stackoverflow.com/questions/21443865/scikit-learn-cross-validation-negative-values-with-mean-squared-error
# scores which are actually losses are negated, actual MSE is the positive part of the score
```

```python
score1 = cross_val_score(regressordt, X, y, cv=10)
print(score1) # with cross validation = 10, scores are computed for each validation and so an array of 10 scores, final score can be computed by average of the array.
import numpy as np
final_score1 = np.average(score1)
print(final_score1)
```
