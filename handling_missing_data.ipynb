```python
#simplest approach for dealing with missing values is to remove entire predictor(s) and/or sample(s) that contain missing values.
df_new = df.dropna()#drop rows with missing values
df_new
```

```python
#Method 2. impute missing values i.e. replace missing values by
#a. a constant value that has meaning in the domain but is distint from all other values.
#b. mean, mode, median of the column
#c. a random value from some other column
#d. a value estimated by another model
df_new2 = df.fillna(df.mean())
print(df_new2)
```

```python
#Method 3. Using SimpleImputer from sklearn
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer()
imputer
```

```python
values = df_new2.values
print(type(values),values.shape, values)
```

```python
imputed_values = imputer.fit_transform(values)
print(type(imputed_values),imputed_values.shape,imputed_values)
```

```python
import numpy as np
print(np.isnan(imputed_values).sum())
```
