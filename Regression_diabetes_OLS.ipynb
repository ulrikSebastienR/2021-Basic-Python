```python
#https://scikit-learn.org/stable/auto_examples/linear_model/plot_ols.html#sphx-glr-auto-examples-linear-model-plot-ols-py

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

diabetes = datasets.load_diabetes
print(diabetes)

diabetes1 = datasets.load_diabetes()
#print(diabetes)
print(type(diabetes1))

for i in diabetes1:
    print(i)

df = pd.DataFrame(diabetes1.data,columns=diabetes1.feature_names)
df["target"] = diabetes1.target

print(df.head())
print(diabetes1.feature_names)

diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
print("diabetes_X= {}, \n, diabetes_y= {}".format(diabetes_X,diabetes_y))

print("diabetes_X.shape = {}, diabetes_y.shape = {}".format(diabetes_X.shape, diabetes_y.shape))

# using np.newaxis to increase dimensions as diabetes_X and diabetes_y have a shape mismatch.

diabetes_X = diabetes_X[:,np.newaxis,2]
print(diabetes_X)
print(diabetes_X.shape)

diabetes_X_train = diabetes_X[:-20]
diabetes_y_train = diabetes_y[:-20]

diabetes_X_test = diabetes_X[-20:]
diabetes_y_test = diabetes_y[-20:]

print(diabetes_X_train, diabetes_y_test)
print(diabetes_X_train.shape, diabetes_y_test.shape)

regr = linear_model.LinearRegression()
print(regr)

regr.fit(diabetes_X_train, diabetes_y_train)

diabetes_y_pred = regr.predict(diabetes_X_test)
print(diabetes_y_pred)

print(regr.coef_)
print(mean_squared_error(diabetes_y_test,diabetes_y_pred))
print(r2_score(diabetes_y_test,diabetes_y_pred))

plt.scatter(diabetes_X_test, diabetes_y_test, color = "black")
plt.plot(diabetes_X_test, diabetes_y_pred, color = "blue", linewidth=3)

plt.xticks()
plt.yticks()

plt.show()

```
