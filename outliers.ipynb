```python
import numpy as np
```

```python
np.random.seed(1)
data = np.random.randn(10000) + 50
print(type(data), data.shape, data)
```

```python
mean = np.mean(data)
std = np.std(data)
print(mean)
print("std", np.std(data))
```

```python
#Finding outliers using Standard Deviation
cutoff = 3*std
x = 0
lower_limit, upper_limit = mean-cutoff, mean+cutoff
print(lower_limit, upper_limit)
```

```python
data_after_outliersremoval = []
for x in data:
  if lower_limit <x <upper_limit:
    data_after_outliersremoval.append(x)

print(type(data_after_outliersremoval), len(data_after_outliersremoval), data_after_outliersremoval)
```
